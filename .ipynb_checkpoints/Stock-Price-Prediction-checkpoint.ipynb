### Statistically Testing for Stationarity

```python
def test_stationarity (timeseries):
    
    '''Function to stattistically test for stationarity in timeseries data
    
    Input: Timeseries data
    Output: 1. Graphical representation of the rolling mean, rolling sd on the original time series plot
            2. Results of Dickey-Fuller test
    
    '''
    
    #Determine rolling statistics
    rolling_mean = pd.Series(timeseries).rolling(window = 20).mean()
    rolling_sd = pd.Series(timeseries).rolling(window = 20).std()
    
    #Plot rolling statistics
    original = plt.plot(timeseries, color = 'blue', label = 'Original')
    mean = plt.plot(rolling_mean, color = 'red', label = 'Rolling mean')
    sd = plt.plot(rolling_sd, color = 'green', label = 'Rolling SD')
    plt.legend(loc = 'best')
    plt.title('Rolling Statistics')
    plt.show()
    
    #Performing Unit Root Tests (Augumented Dickey-Fuller Test)
    print('Results of Augumented Dickey-fuller Test:')
    df_test = adfuller(timeseries)
    print('ADF Statistic: %f' % df_test[0])
    print('p-value: %f' % df_test[1])
    print('Critical Values:')
    for key, value in df_test[4].items():
        print('\t%s: %.3f' % (key, value))
    
```

```python
test_stationarity(ts)
```

We interpret this result using the p-value from the test.

p-value > 0.05: Fail to reject the null hypothesis (H0), the data has a unit root and is non-stationary.

p-value <= 0.05: Reject the null hypothesis (H0), the data does not have a unit root and is stationary.

# Making the data stationary

```python
#Estimating & eliminating the trend
ts_log = np.log(ts)
plt.plot(ts_log)
plt.show()
```

#### Smoothing with Moving Average

```python
moving_avg = pd.Series(ts_log).rolling(window = 10).mean()
plt.plot(ts_log)
plt.plot(moving_avg, color = 'red')
plt.show()
```

```python
#viewing a sample of the moving average
moving_avg.sample(5)
```

```python
#differentiating
ts_diff = ts_log - moving_avg
ts_diff.sample(5)
```

```python
#dropping missing values
ts_diff.dropna(inplace = True)
ts_diff.head()
```

```python
#re-checking for stationarity in the transformed data
test_stationarity(ts_diff)
```

Now the p-value is < 0.05 so we reject the null hypothesis and conclude that the data is stationary  

```python
#Performing decomposition
decomposition = seasonal_decompose(ts_log, period = 52)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(ts_log, label='Observed')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

```python
##re-checking for stationarity in the decomposed data
ts_log_decompose = residual
ts_log_decompose.dropna(inplace=True)
test_stationarity(ts_log_decompose)
```

Here the p-value is < 0.05 so we reject the null hypothesis and conclude that the data is stationary.
